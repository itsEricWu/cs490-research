```python
val = pickle.load(open("/home/lu677/cs490/cs490-research/Resnet/nesterov/l_val_loss", "rb"))
train = pickle.load(open("/home/lu677/cs490/cs490-research/Resnet/nesterov/l_train_loss", "rb"))
plt.bar(["with nesterov", "without nesterov"], train, label="Training Loss")
plt.xlabel("nesterov momentum")
plt.ylabel("loss")
plt.title("training with nesterov momentum vs loss")
plt.legend()
plt.show()

plt.bar(["with nesterov", "without nesterov"], val, label="Validation Loss")
plt.xlabel("nesterov momentum")
plt.ylabel("loss")
plt.title("validation with nesterov momentum vs loss")
plt.legend()
plt.show()


# Numbers of pairs of bars you want
N = 2

# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = (val)
# Specify the values of orange bars (height)
orange_bar = (train)

# Position of bars on x-axis
ind = np.arange(N)


# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, blue_bar , width, label='validation loss')
plt.bar(ind + width, orange_bar, width, label='train loss')

plt.xlabel('nesterov momentum')
plt.ylabel('loss')
plt.title('nestrov momentum vs loss')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, ('with nesterov', 'without nesterov'))

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```
